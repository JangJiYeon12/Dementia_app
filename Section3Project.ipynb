```python
import pandas as pd
import numpy as np
import io
```

```python
train_x = pd.read_csv('/content/train.csv')
train_y = pd.read_csv('/content/train_label.csv', index_col=0)
```

```python
train_x
```

```python
#drop colnums

train_x1 = train_x[['EMAIL', 'activity_steps', 'activity_high', 'sleep_duration', 'sleep_light', 'sleep_midpoint_time', 'sleep_deep', 'sleep_restless', 'sleep_score', 'sleep_hr_average']]
```

```python
train_x1
```

```python
#groupby by EMAIL

train_x2 = train_x1.groupby(['EMAIL'], as_index=True).mean()
```

```python
train_x2
```

```python
size_mapping = {'CN':0,'MCI':1,'Dem':2}
train_y['DIAG_NM'] = train_y['DIAG_NM'].map(size_mapping)

train_y
```

```python
data = pd.merge(train_x2, train_y, left_index=True, right_index=True)
```

```python
data.round(0)
```

```python
data = data.astype(int)
```

```python
data
```

```python
 def to_min(x):
     x = x / 60
     return round(x)
```

```python
sleep_data = data.filter(regex='sleep_')

sleep_hr = sleep_data[['sleep_restless', 'sleep_hr_average', 'sleep_score']]

sleep_data = sleep_data.drop(['sleep_restless', 'sleep_hr_average', 'sleep_score'], axis = 1)

activity_data = data.filter(regex='activity_')
```

```python
sleep_data = sleep_data.apply(to_min).astype(int)
```

```python
df1 = pd.merge(activity_data, sleep_data, left_index=True, right_index=True)
```

```python
df1 = pd.merge(df1, sleep_hr, left_index=True, right_index=True)
```

```python
df1
```

```python
from sklearn.model_selection import train_test_split

target = 'DIAG_NM'
features = df1.columns

X_train, X_test, y_train, y_test = train_test_split(df1, train_y, stratify=train_y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import model_selection
from sklearn import metrics

model = LogisticRegression(solver='liblinear') 
```

```python
model.fit(X_train, y_train)
```

```python
y_pred_ord = model.predict(X_test)
```

```python
print('TrainData Accuracy : ', "{:.5f}".format(model.score(X_train, y_train)))
print('TestData Accuracy : ', "{:.5f}".format(model.score(X_test, y_test)))
```

```python
import pickle
filename = 'finalized_model.pkl'
pickle.dump(model, open(filename, 'wb'))
```
